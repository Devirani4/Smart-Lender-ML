# Univariate Analysis

Univariate analysis is used to study one feature (column) at a time.

Continuous variables are visualized using Distribution Plots.

Categorical variables are visualized using Count Plots.

```python
# Distribution Plot

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.distplot(data['ApplicantIncome'], color='red')

plt.subplot(1,2,2)
sns.distplot(data['Credit_History'])

plt.show()
```

```python
# Count Plot

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.countplot(x='Gender', data=data)

plt.subplot(1,2,2)
sns.countplot(x='Education', data=data)

plt.show()
```

```python
cat = [
    'Gender',
    'Married',
    'Dependents',
    'Education',
    'Self_Employed',
    'Property_Area',
    'Loan_Status'
]

plt.figure(figsize=(18,12))

for i, column in enumerate(cat, 1):

    plt.subplot(3,3,i)

    sns.countplot(x=column, data=data)

plt.tight_layout()

plt.show()
```

## Observations

- ApplicantIncome has a left-skewed distribution.
- Credit_History contains binary values (0 and 1).
- Gender has two categories: Male and Female.
- Education has two categories: Graduate and Not Graduate.
- Most applicants are Male.
- Most applicants are Graduates.

# Bivariate Analysis

Bivariate analysis is used to study the relationship between two variables.

Count plots are used to compare categorical features and identify patterns between them.

```python
# Visualising two columns against each other

plt.figure(figsize=(20,5))

# Married vs Gender
plt.subplot(1,3,1)
sns.countplot(x='Married', hue='Gender', data=data)

# Self Employed vs Education
plt.subplot(1,3,2)
sns.countplot(x='Self_Employed', hue='Education', data=data)

# Property Area vs Loan Amount Term
plt.subplot(1,3,3)
sns.countplot(x='Property_Area', hue='Loan_Amount_Term', data=data)

plt.tight_layout()
plt.show()
```

## Observations

- The relationship between Gender and Married status can be observed using the first count plot.
- Most married applicants are male.
- Graduate applicants are more common among both self-employed and non-self-employed groups.
- Loan amount terms vary across different property areas.
- Bivariate analysis helps understand relationships between two variables before building the machine learning model.

# Multivariate Analysis

Multivariate analysis is used to study the relationship between multiple variables simultaneously.

In this analysis, Gender, ApplicantIncome and Loan_Status are visualized using a Swarm Plot.

```python
# Visualizing Gender, Applicant Income and Loan Status

plt.figure(figsize=(10,6))

sns.swarmplot(
    x='Gender',
    y='ApplicantIncome',
    hue='Loan_Status',
    data=data
)

plt.title("Gender vs Applicant Income by Loan Status")

plt.show()
```

## Observations

- The swarm plot visualizes the relationship between Gender, ApplicantIncome and Loan_Status.
- Applicants with different income levels are distributed across both genders.
- Loan approval and rejection are represented using different colors.
- This analysis helps identify patterns among multiple variables simultaneously.

Data Pre-Processing:

1.Checking and Handling Missing Values

```python
# Check missing values
print(data.isnull().sum())
```

```python
# Fill categorical columns with mode
data['Gender'].fillna(data['Gender'].mode()[0], inplace=True)
data['Married'].fillna(data['Married'].mode()[0], inplace=True)
data['Dependents'].fillna(data['Dependents'].mode()[0], inplace=True)
data['Self_Employed'].fillna(data['Self_Employed'].mode()[0], inplace=True)
data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mode()[0], inplace=True)
data['Credit_History'].fillna(data['Credit_History'].mode()[0], inplace=True)

# Fill numerical column with median
data['LoanAmount'].fillna(data['LoanAmount'].median(), inplace=True)

# Verify
print(data.isnull().sum())
```

2. Balancing the Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Loan_Status', data=data)
plt.title("Loan Status Distribution")
plt.show()

print(data['Loan_Status'].value_counts())
```

3. Scaling the Data

```python
# Encode categorical columns

data['Gender'] = data['Gender'].map({'Male': 1, 'Female': 0})
data['Married'] = data['Married'].map({'Yes': 1, 'No': 0})
data['Dependents'] = data['Dependents'].replace('3+', 3)
data['Dependents'] = data['Dependents'].astype(int)
data['Education'] = data['Education'].map({'Graduate': 1, 'Not Graduate': 0})
data['Self_Employed'] = data['Self_Employed'].map({'Yes': 1, 'No': 0})
data['Property_Area'] = data['Property_Area'].map({
    'Rural': 0,
    'Semiurban': 1,
    'Urban': 2
})
```

```python
X = data.drop(['Loan_ID', 'Loan_Status'], axis=1)
y = data['Loan_Status']
```

```python
print(X.select_dtypes(include='object').columns)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

4. Splitting Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

print("Training Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)
```
